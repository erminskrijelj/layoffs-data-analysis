```python
mydb = mysql.connector.connect(
  host = host,
  user = username,
  password = password,
  database="word_layoffs"
)
```

```python

mycursor = mydb.cursor()

mycursor.execute("SELECT * FROM layoffs")

myresult = mycursor.fetchall()

mydict=dict()
```

```python

for i,x in enumerate(myresult):
  mydict[i] = x
```

```python
mycursor.execute("DESCRIBE layoffs")

myresult = mycursor.fetchall()
columns_names = [] 
```

```python
for res in myresult:
    columns_names.append(res[0]) 
```
